```python
from transformers import pipeline
from pydub import AudioSegment, silence
import io, zipfile
from google.colab import files

# 1ï¸âƒ£ Audioni yuklash
audio_path = "/content/Kutmang, kutish og'riq beradi ï½œ Iqtibos podcast #4_temp.wav"  # Sizning wav faylingiz
audio = AudioSegment.from_wav(audio_path)

# 2ï¸âƒ£ Gap tugashiga qarab boâ€˜lish
chunks = silence.split_on_silence(
    audio,
    min_silence_len=500,  # 0.5s jimlik
    silence_thresh=-40,
    keep_silence=200
)

# 3ï¸âƒ£ 30 sekunddan uzun boâ€˜laklarni 30 sekundgacha boâ€˜lish
final_chunks = []
for chunk in chunks:
    if len(chunk) > 30*1000:
        start = 0
        while start < len(chunk):
            end = min(start + 30*1000, len(chunk))
            final_chunks.append(chunk[start:end])
            start = end
    else:
        final_chunks.append(chunk)

print(f"âœ… {len(final_chunks)} boâ€˜lak yaratildi")

# 4ï¸âƒ£ MMS-1b-all modelini yuklash
asr = pipeline("automatic-speech-recognition", model="facebook/mms-1b-all")

# 5ï¸âƒ£ Zip fayl yaratish
zip_buffer = io.BytesIO()
with zipfile.ZipFile(zip_buffer, "w") as zipf:
    for i, chunk in enumerate(final_chunks):
        # Audio boâ€˜lakni wav ga yozish
        wav_io = io.BytesIO()
        chunk.export(wav_io, format="wav")
        wav_name = f"chunk_{i+1}.wav"
        zipf.writestr(wav_name, wav_io.getvalue())

        # Audio boâ€˜lakni matnga aylantirish
        wav_io.seek(0)
        result = asr(wav_io.read())
        text_name = f"chunk_{i+1}.txt"
        zipf.writestr(text_name, result['text'])

# 6ï¸âƒ£ Zip faylni Colabâ€™dan yuklab olish
zip_buffer.seek(0)
with open("audio_texts.zip", "wb") as f:
    f.write(zip_buffer.read())

files.download("audio_texts.zip")
```

```python
from transformers import pipeline
from pydub import AudioSegment, silence
import io, zipfile
from google.colab import files

# 1ï¸âƒ£ Audioni yuklash
audio_path = "/content/Odatlar sizni yo'q qiladi va quradi ï½œ  Iqtibos podcast #2_temp.wav"  # fayl nomini soddalashtirdim
audio = AudioSegment.from_wav(audio_path)

# 2ï¸âƒ£ Gap tugashiga qarab boâ€˜lish
chunks = silence.split_on_silence(
    audio,
    min_silence_len=500,   # 0.5s jimlik
    silence_thresh=-40,
    keep_silence=200
)

# 3ï¸âƒ£ 30 sekunddan uzun boâ€˜laklarni boâ€˜lish
final_chunks = []
for chunk in chunks:
    if len(chunk) > 30*1000:
        start = 0
        while start < len(chunk):
            end = min(start + 30*1000, len(chunk))
            final_chunks.append(chunk[start:end])
            start = end
    else:
        final_chunks.append(chunk)

print(f"âœ… {len(final_chunks)} boâ€˜lak yaratildi")

# ðŸ”§ Qisqa boâ€˜laklarga moslash
MIN_DURATION_MS = 1000  # 1 soniyadan kichik boâ€˜lsa
def pad_audio(seg, min_ms=MIN_DURATION_MS):
    """Agar boâ€˜lak juda qisqa boâ€˜lsa, oxiriga jimlik qoâ€˜shadi"""
    if len(seg) < min_ms:
        silence_seg = AudioSegment.silent(duration=min_ms - len(seg))
        seg = seg + silence_seg
    return seg

# 4ï¸âƒ£ MMS-1b-all modelini yuklash
asr = pipeline("automatic-speech-recognition", model="facebook/mms-1b-all")

# 5ï¸âƒ£ Zip fayl yaratish
zip_buffer = io.BytesIO()
with zipfile.ZipFile(zip_buffer, "w") as zipf:
    for i, chunk in enumerate(final_chunks):
        # ðŸ”§ Qisqa boâ€˜laklarni moslashtirish
        chunk = pad_audio(chunk)

        # Audio boâ€˜lakni wav ga yozish
        wav_io = io.BytesIO()
        chunk.export(wav_io, format="wav")
        wav_name = f"chunk_{i+1}.wav"
        zipf.writestr(wav_name, wav_io.getvalue())

        # Audio boâ€˜lakni matnga aylantirish
        wav_io.seek(0)
        result = asr(wav_io.read())
        text_name = f"chunk_{i+1}.txt"
        zipf.writestr(text_name, result['text'])

# 6ï¸âƒ£ Zip faylni Colabâ€™dan yuklab olish
zip_buffer.seek(0)
with open("audio_texts.zip", "wb") as f:
    f.write(zip_buffer.read())

files.download("audio_texts.zip")
```
