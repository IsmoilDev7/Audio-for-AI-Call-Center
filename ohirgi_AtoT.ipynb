```python
from transformers import pipeline
from pydub import AudioSegment, silence
import io, zipfile
from google.colab import files

# 1️⃣ Audioni yuklash
audio_path = "/content/Kutmang, kutish og'riq beradi ｜ Iqtibos podcast #4_temp.wav"  # Sizning wav faylingiz
audio = AudioSegment.from_wav(audio_path)

# 2️⃣ Gap tugashiga qarab bo‘lish
chunks = silence.split_on_silence(
    audio,
    min_silence_len=500,  # 0.5s jimlik
    silence_thresh=-40,
    keep_silence=200
)

# 3️⃣ 30 sekunddan uzun bo‘laklarni 30 sekundgacha bo‘lish
final_chunks = []
for chunk in chunks:
    if len(chunk) > 30*1000:
        start = 0
        while start < len(chunk):
            end = min(start + 30*1000, len(chunk))
            final_chunks.append(chunk[start:end])
            start = end
    else:
        final_chunks.append(chunk)

print(f"✅ {len(final_chunks)} bo‘lak yaratildi")

# 4️⃣ MMS-1b-all modelini yuklash
asr = pipeline("automatic-speech-recognition", model="facebook/mms-1b-all")

# 5️⃣ Zip fayl yaratish
zip_buffer = io.BytesIO()
with zipfile.ZipFile(zip_buffer, "w") as zipf:
    for i, chunk in enumerate(final_chunks):
        # Audio bo‘lakni wav ga yozish
        wav_io = io.BytesIO()
        chunk.export(wav_io, format="wav")
        wav_name = f"chunk_{i+1}.wav"
        zipf.writestr(wav_name, wav_io.getvalue())

        # Audio bo‘lakni matnga aylantirish
        wav_io.seek(0)
        result = asr(wav_io.read())
        text_name = f"chunk_{i+1}.txt"
        zipf.writestr(text_name, result['text'])

# 6️⃣ Zip faylni Colab’dan yuklab olish
zip_buffer.seek(0)
with open("audio_texts.zip", "wb") as f:
    f.write(zip_buffer.read())

files.download("audio_texts.zip")
```

```python
from transformers import pipeline
from pydub import AudioSegment, silence
import io, zipfile
from google.colab import files

# 1️⃣ Audioni yuklash
audio_path = "/content/Odatlar sizni yo'q qiladi va quradi ｜  Iqtibos podcast #2_temp.wav"  # fayl nomini soddalashtirdim
audio = AudioSegment.from_wav(audio_path)

# 2️⃣ Gap tugashiga qarab bo‘lish
chunks = silence.split_on_silence(
    audio,
    min_silence_len=500,   # 0.5s jimlik
    silence_thresh=-40,
    keep_silence=200
)

# 3️⃣ 30 sekunddan uzun bo‘laklarni bo‘lish
final_chunks = []
for chunk in chunks:
    if len(chunk) > 30*1000:
        start = 0
        while start < len(chunk):
            end = min(start + 30*1000, len(chunk))
            final_chunks.append(chunk[start:end])
            start = end
    else:
        final_chunks.append(chunk)

print(f"✅ {len(final_chunks)} bo‘lak yaratildi")

# 🔧 Qisqa bo‘laklarga moslash
MIN_DURATION_MS = 1000  # 1 soniyadan kichik bo‘lsa
def pad_audio(seg, min_ms=MIN_DURATION_MS):
    """Agar bo‘lak juda qisqa bo‘lsa, oxiriga jimlik qo‘shadi"""
    if len(seg) < min_ms:
        silence_seg = AudioSegment.silent(duration=min_ms - len(seg))
        seg = seg + silence_seg
    return seg

# 4️⃣ MMS-1b-all modelini yuklash
asr = pipeline("automatic-speech-recognition", model="facebook/mms-1b-all")

# 5️⃣ Zip fayl yaratish
zip_buffer = io.BytesIO()
with zipfile.ZipFile(zip_buffer, "w") as zipf:
    for i, chunk in enumerate(final_chunks):
        # 🔧 Qisqa bo‘laklarni moslashtirish
        chunk = pad_audio(chunk)

        # Audio bo‘lakni wav ga yozish
        wav_io = io.BytesIO()
        chunk.export(wav_io, format="wav")
        wav_name = f"chunk_{i+1}.wav"
        zipf.writestr(wav_name, wav_io.getvalue())

        # Audio bo‘lakni matnga aylantirish
        wav_io.seek(0)
        result = asr(wav_io.read())
        text_name = f"chunk_{i+1}.txt"
        zipf.writestr(text_name, result['text'])

# 6️⃣ Zip faylni Colab’dan yuklab olish
zip_buffer.seek(0)
with open("audio_texts.zip", "wb") as f:
    f.write(zip_buffer.read())

files.download("audio_texts.zip")
```

```python
from transformers import pipeline
from pydub import AudioSegment, silence
import io, zipfile
from google.colab import files

audio_path = "/content/URIK GULI： MEN O_ZBEKMAN!_temp.wav"
audio = AudioSegment.from_wav(audio_path)

chunks = silence.split_on_silence(
    audio,
    min_silence_len=500,
    silence_thresh=-40,
    keep_silence=200
)

final_chunks = []
buffer_chunk = AudioSegment.empty()

for chunk in chunks:
    if len(chunk) > 30*1000:
        start = 0
        while start < len(chunk):
            end = min(start + 30*1000, len(chunk))
            final_chunks.append(chunk[start:end])
            start = end
    else:
        buffer_chunk += chunk
        if len(buffer_chunk) >= 10*1000:
            final_chunks.append(buffer_chunk)
            buffer_chunk = AudioSegment.empty()


if len(buffer_chunk) > 0:
    final_chunks.append(buffer_chunk)

print(f"✅ {len(final_chunks)} bo‘lak yaratildi")


asr = pipeline("automatic-speech-recognition", model="facebook/mms-1b-all")


zip_buffer = io.BytesIO()
with zipfile.ZipFile(zip_buffer, "w") as zipf:
    for i, chunk in enumerate(final_chunks):

        wav_io = io.BytesIO()
        chunk.export(wav_io, format="wav")
        wav_name = f"chunk_{i+1}.wav"
        zipf.writestr(wav_name, wav_io.getvalue())


        wav_io.seek(0)
        result = asr(wav_io.read())
        text_name = f"chunk_{i+1}.txt"
        zipf.writestr(text_name, result['text'])


zip_buffer.seek(0)
with open("audio_texts.zip", "wb") as f:
    f.write(zip_buffer.read())

files.download("audio_texts.zip")
```
